```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
from matplotlib import *
import sys
from pylab import *
#Libraries we will use
```

```python
data1 = pd.read_csv("C:/Users/USER/Desktop/accidents_2005_to_2007.csv")
data2 = pd.read_csv("C:/Users/USER/Desktop/accidents_2009_to_2011.csv")
data3 = pd.read_csv("C:/Users/USER/Desktop/accidents_2012_to_2014.csv")

## Read Data. data1 = accidents 2005 to 2007, data2 = accidents 2009 to 2011, data3 =  accidents 2012 to 2014 
```

```python
dataAll = pd.concat([data1,data2,data3]) #Merge all data
```

```python
dataNew = dataAll.loc[:, ["Accident_Severity", "Date", "Day_of_Week", "Time", "Speed_limit", 
                        "Road_Type", "Light_Conditions", "Weather_Conditions", "Road_Surface_Conditions",
                        "Carriageway_Hazards", "Urban_or_Rural_Area", "Year",
                        "Number_of_Vehicles", "Number_of_Casualties"]]

#we only get data that we will use
```

```python
dataNew.info() #information about the data we extract
```

```python
dataNew.isnull().sum() #Checking the null values of the data we extracted
```

```python
dataNew.dropna(inplace=True)
dataNew.shape
#null value inference
```

```python
dataNew.describe() #descriptive statistical values of the data
```

```python
dataNew
```

```python
plt.figure(figsize=(22,10))
sns.countplot(x="Year",data=dataNew)
plt.title("Number of Accidents by Year")
plt.xlabel("Years")
plt.ylabel("Number of Accidents")
#Number of accidents by year
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(22,10))
sns.countplot(x="Day_of_Week",data=dataNew)
plt.title("Number Of Accidents by Days Of Week")
plt.xlabel("Days Of Week")
plt.ylabel("Numbers Of Accidents")
#Number of accidents by day of the week
```

```python
hours = pd.DataFrame(dataNew["Time"].astype("datetime64"))
hours = pd.DataFrame(hours.Time.dt.strftime('%H').sort_values(),columns=["Time"])

plt.figure(figsize=(22,10))
sns.countplot(data=hours,x="Time")
plt.title("Number Of Accidents by Hours")
plt.xlabel("Hours")
plt.ylabel("Number Of Accidents")
#number of accidents per hour
```

```python
plt.figure(figsize=(22,10))
sns.countplot(data=dataNew,x="Road_Surface_Conditions",hue="Year")
# Number of accidents by years depending on road conditions
```

```python
plt.figure(figsize=(22,10))
sns.countplot(data=dataNew,x="Road_Type",hue="Accident_Severity")
plt.title("Accident Severity By Road Type",size=20)
plt.xlabel("Road Type",size=20)
plt.ylabel("Numbers Of Accidents",size=20)
plt.legend(title="Accident Severity",labels=["Fatal","Serious","Slight"])
# road type and severity of the accident
```

```python
plt.figure(figsize=(22,10))
sns.countplot(data=dataNew,x="Weather_Conditions",hue="Accident_Severity")
plt.title("Accident Severity By Weather Conditions",size=20)
plt.xlabel("Weather Conditions",size=20)
plt.ylabel("Numbers Of Accidents",size=20)
plt.legend(title="Accident Severity",labels=["Fatal","Serious","Slight"])
#weather conditions and the severity of the accident
```

```python
plt.figure(figsize=(22,10))
sns.countplot(data=dataNew,x="Speed_limit",hue="Accident_Severity")
plt.title("Accident Severity By Speed Limit",size=20)
plt.xlabel("Speed Limit",size=20)
plt.ylabel("Numbers Of Accidents",size=20)
plt.legend(title="Accident Severity",labels=["Fatal","Serious","Slight"])
#speed limit and accident severity
```

```python
plt.figure(figsize=(22,10))
sns.countplot(data=dataNew,x="Light_Conditions",hue="Accident_Severity")
plt.title("Accident severity according to Light Conditions",size=20)
plt.xlabel("Light Conditions",size=20)
plt.ylabel("Numbers Of Accidents",size=20)
plt.legend(title="Accident Severity",labels=["Fatal","Serious","Slight"])
#light conditions accident severity
```

```python
area = pd.DataFrame(dataNew["Urban_or_Rural_Area"].value_counts())
area = area.drop([3])

area.plot.pie(y="Urban_or_Rural_Area",figsize=[10,10])
plt.title("Accidents of Urban Or Rural")
plt.legend(title="Accidents of Urban Or Rural",labels=["Urban","Area"])
#urban and rural accident rates
```
